```python
tensor = pd.read_pickle("data-sim/cc/train_weighted_labels.pkl")
```

```python
def show_row(tensor, row):
    row_data_np = tensor[row].numpy()
    filtered_data = row_data_np[row_data_np != 0]
    filtered_data = filtered_data[filtered_data != 1]
    min_ = min(filtered_data)
    max_ = max(filtered_data)
    # Plot the histogram
    plt.hist(filtered_data, bins=500, alpha=0.5, color='g', edgecolor='black')
    plt.title(f"Histogram for Row {row}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()
    return min_, max_
```

```python

for row in range(len(tensor)):
    gt_0 = (tensor[row] > 0).float()
    lt_1 = (tensor[row] < 1).float()
    both = (gt_0 * lt_1).sum()
    if both > 0:
        min_, max_ = show_row(tensor, row)
        print(f"Found at {row}: {both}. Min: {min_}. Max: {max_}")

        #break
```
